## Unnecessary Data

Look for the different types of unnecessary data in your dataset and address it as needed. Make sure to use code comments to illustrate your thought process.

Here's where things get interesting... we're going to first remove all the columns we don't need. As a reminder, the different columns are listed below.

```python
list(df_mega.columns)
print(df_mega.shape)
df_mega = df_mega.drop(columns = ['Adj. O',
 'Adj. D',
 'T',
 'O.EFF',
 'O.eFG%',
 'O.TO%',
 'O.Reb%',
 'O.FTR',
 'D.EFF',
 'D.eFG%',
 'D.TO%',
 'D.Reb%',
 'D.FTR',
 'G-Sc',
 '+/-'])
df_mega.head(5)
```

Additionally, during EDA I discovered that there are many games that are not relevant for my analysis. These are the conference and conference tournament games. I will now go ahead and remove those datapoints as well.

```python
# Seeing what the different types of 'Type', or game type, there are.
print(df_mega.Type.unique())
print(df_mega.Venue.unique())
# We are removing the 'C' and 'CT' values, as these stand for conference and conference tournament.
filter_values = ['N','T','P']
df_mega = df_mega[df_mega['Type'].isin(filter_values)]
print(df_mega.shape)
df_mega.info
```

That's much easier to look at and work with! During my EDA portion of the final project, I ran into an issue where each game is in the database twice. Below I will try and rectify that issue. There are 40608 rows now, so that means that I should have 20304 rows once duplicates are removed below.

```python
# Creating a unique game ID for each game... this should help eliminate the duplicates.
df_mega["game_id"] = (
    df_mega[["Team","Opp."]].apply(lambda x: "_".join(sorted(x)), axis = 1)
)

# Based on some research I did, the best way to get the opponent conference is to "self-merge" from the duplicate row. Code below.
df_mega = df_mega.merge(
    df_mega[["game_id", "Date", "Team", "Conf."]],
    left_on = ["game_id", "Date","Opp."],
    right_on = ["game_id", "Date", "Team"],
    how = "left",
    suffixes = ("","_opp")
)

df_mega = df_mega.rename(columns = {"Conf._opp":"Opp_Conf"})

# We're also going to preserve the neutral-site games, because we're ultimately going to remove the remainder of the away games
venue_rank = {
    "Home": 0,
    "Neutral" : 1,
    "Away": 2
}

df_mega['venue_rank'] = df_mega["Venue"].map(venue_rank)

```

```python
# Finally, we'll drop the duplicate rows!
df_mega = df_mega.sort_values("venue_rank").drop_duplicates(subset=["game_id","Date"]).drop(columns=["venue_rank","game_id","Team_opp"])
```
